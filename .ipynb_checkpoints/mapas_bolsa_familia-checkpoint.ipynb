## 2 - Pré-processamento dos dados

Nesta seção, vamos formatar os dados para que estejam prontos para visualização no mapa do Brasil.

Antes, vamos fazer uma amostra simples dos dados de pagamentos e saques, pois temos muitos dados para processar (ainda que referentes apenas ao mês de junho de 2017!). Vamos fazer uma amostra de 10% da população para as tabelas de pagamentos e saques: 

```r
tamanho_amostra <- 0.1*nrow(pagamentos)
indice_amostra_pag <- sample(nrow(pagamentos), size = tamanho_amostra)
amostra_pagamentos <- pagamentos[indice_amostra_pag,]

tamanho_amostra <- 0.1*nrow(saques)
indice_amostra_saques <- sample(nrow(saques), size = tamanho_amostra)
amostra_saques <- saques[indice_amostra_saques,]
```

Agora podemos esquecer as tabelas completas de pagamentos e saques e vamos mexer apenas nas amostras. 

O primeiro problema que temos é que os campos de valor das amostras de pagamentos e saques estão como tipo texto (character). Primeiro, vamos mudar os campos desses valores para numérico.

```r
amostra_pagamentos$Valor.Parcela <- as.numeric(amostra_pagamentos$Valor.Parcela)
amostra_saques$Valor.Parcela <- as.numeric(amostra_saques$Valor.Parcela)
```

Agora vamos agregar a amostra de pagamentos por município, código de município e UF e somar os valores pagos. Podemos fazer isso através das funções da biblioteca dplyr.

```r
library(dplyr) # rodar install.packages('dplyr'), caso a biblioteca não esteja instalada
pagamentos_agregados <- amostra_pagamentos %>% group_by(Nome.Município, Código.SIAFI.Município, UF) %>%  summarise(Valor.Parcela = sum(Valor.Parcela)) 
```

```r
pagamentos_agregados
```

Agora façamos o mesmo para os saques...

```r
saques_agregados <- amostra_saques %>% group_by(Nome.Município, Código.SIAFI.Município, UF) %>%  summarise(Valor.Parcela = sum(Valor.Parcela)) 
```

```r
saques_agregados
```

Agora vamos unir as duas tabelas para calcular o valor total em cada município. Podemos fazer isso através da função merge do R.

```r
total_bolsa_familia <- merge(pagamentos_agregados, saques_agregados, by = c('Nome.Município','Código.SIAFI.Município','UF'))
```

```r
head(total_bolsa_familia)
```

Finalmente, somamos o valor das parcelas para ter o valor total.

```r
total_bolsa_familia$ValorTotal = total_bolsa_familia$Valor.Parcela.x + total_bolsa_familia$Valor.Parcela.y
```

```r
head(total_bolsa_familia)
```

A tarefa final é juntar os valores do bolsa família com os dados de polígonos dos municípios brasileiros. Podemos fazer isso novamente através do comando merge usando as colunas de UF e município.

```r
merge(municipios@data, total_bolsa_familia, by.x = c('NOME_MUNI','UF'), by.y = c('Nome.Município', 'UF'))
```

Veja que o merge retornou uma tabela vazia. Isso aconteceu porque estamos comparando nomes de municípios com letras maiúsculas e sem acento (na tabela de valores totais do Bolsa Família) com municípios com letras minúsculas e com acento (na tabela de polígonos dos municípios). Para resolvermos este problema, vamos transformar os nomes de municípios da tabela de polígonos para letras maiúsculas e sem acento.

```r
municipios@data$NOME_MUNI <- toupper(municipios@data$NOME_MUNI)
municipios@data$NOME_MUNI <- iconv(municipios@data$NOME_MUNI,to="ASCII//TRANSLIT")
head(municipios@data$NOME_MUNI)
```

Agora podemos aplicar a união destas tabelas.

## 3 - Integração dos dados do Bolsa Família com os dados dos municípios

A integração entre as tabelas do Programa e dos Municípios é feita novamente através do comando merge.

```r
municipios@data <- merge(municipios@data, total_bolsa_familia, by.x = c('NOME_MUNI','UF'), by.y = c('Nome.Município', 'UF'))
head(municipios@data)
```

Observe que os valores do bolsa família foram agregados aos dados dos municípios.

## 4 - Construção do Mapa 

Por fim, nesta última seção, construiremos o mapa de distribuição do Bolsa Família por valor total de saques e pagamentos ocorridos em junho de 2017.

Primeiro temos que construir nossa paleta de cores:

```r
paleta.verde <- brewer.pal(9,'Greens')
criarCores  <- colorRampPalette(paleta.verde)
cores  <- criarCores(nrow(municipios@data))
pie(rep(1,nrow(municipios@data)), col=cores, lty=0, labels=NA)
```

Através dos comandos a seguir, criamos uma coluna "ValorTotal.categoria" de modo que cada cor da paleta seja uma categoria de Valor Total gasto com Bolsa Família (do verde mais claro para o verde mais escuro).

```r
numeroCores <- length(unique(cores))
cuts <- quantile(municipios@data$ValorTotal, probs=(0:numeroCores)/numeroCores)
municipios@data$ValorTotal.categoria <- cut(municipios$ValorTotal, breaks = cuts, include.lowest=T)
cor.das.categorias <- data.frame(ValorTotal.categoria=levels(municipios$ValorTotal.categoria),cores=unique(cores))
municipios@data  <- merge(municipios@data,cor.das.categorias)
```

Finalmente, desenhamos o mapa a seguir:

```r
plot(municipios, col=as.character(municipios@data$cores), lty=0, main='Total Gasto com Bolsa Família (junho/2017)')
 
# legend
color.legend(xl=-70, xr=-60,yb=-25,yt=-26, 
         legend=c('lowest','','highest'), 
         rect.col=cores, gradient='x',
         cex=.7, pos=c(1,1,1))
```
